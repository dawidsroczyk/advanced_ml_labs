```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Generate data

```python
N = 2000
p = 0.9

n0 = np.random.binomial(N, p)
n1 = N - n0

X0 = np.random.normal(5, 1, size=n0)
X1 = np.random.normal(10, 1, size=n1)
X = np.concat([X0, X1])
```

```python
sns.histplot(np.concat([X0, X1]), kde=True)
```

```python
def kernel_1(X):
    return (np.abs(X) < 0.5).astype(int)

def kernel_2(X):
    return stats.norm.pdf(X)
```

```python
hn = 0.1
```

```python
x_dens = np.linspace(0, 12)
y_dens = np.sum(kernel_2((x_dens.reshape((-1, 1)) - X) / hn), axis=1) / (N * hn)
y_real = 0.9 * stats.norm.pdf(x_dens, 5, 1) + 0.1 * stats.norm.pdf(x_dens, 10, 1)
plt.plot(x_dens, y_dens)
plt.plot(x_dens, y_real)
```
